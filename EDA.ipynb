Bivariate analysis:
Types of analysis-
1. Numeric-Numeric [scatter plot 2D hist,2D Kde]
2. Categorical-Numeric [Bar,Boxplot,kde]
3. Categorical-Categorical [Heatmaps, Stacked Barchats]

```python
# Categorical - Numeric
#Age-diabetic

pd.crosstab(df['diabetic'],df['age'])
```

```python
pd.crosstab(df['age'],df['diabetic'],normalize = 'columns').head()
```

```python
pd.crosstab(df['age'],df['diabetic'],normalize = 'columns').plot(kind='kde')
```

```python
#Age-smoker

pd.crosstab(df['age'],df['smoker'],normalize = 'columns').head()
```

```python
pd.crosstab(df['age'],df['smoker'],normalize = 'columns').plot(kind = 'kde')
```

```python
pd.crosstab(df['age'],df['smoker'],normalize = 'columns').plot(kind = 'bar')
```
